```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
!pip install openpyxl
```

```python
data = pd.read_excel('applications.xlsx')
print(data)

print(data.isnull().sum())
```

```python
plt.scatter(data['Points'],data['Percentage'])
plt.xlabel('Points')
plt.ylabel('Percentage')
plt.show()
```

```python
km = KMeans(n_clusters = 3)
cluster = km.fit_predict(data[['Points','Percentage']])
print(cluster)
```

```python
data['cluster'] = cluster
data1 = data[data.cluster == 0]
data2 = data[data.cluster == 1]
data3 = data[data.cluster == 2]
print(data1)
print()
print(data2)
print()
print(data3)
```

```python
plt.scatter(data1['Points'],data1['Percentage'],color = 'green')
plt.scatter(data2['Points'],data2['Percentage'],color = 'red')
plt.scatter(data3['Points'],data3['Percentage'],color = 'blue')
```
